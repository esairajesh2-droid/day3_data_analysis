```python
#bar chart
import matplotlib.pyplot as plt
import pandas as pd
data1=pd.DataFrame({
    'India':[10],
    'USA':[25],
    'Japan':[18],
    'Brazil':[8]
})
x=data1.columns
y=data1.iloc[0]
plt.figure(figsize=(8,5))
plt.bar(x,y,color='red',edgecolor='green',linewidth=0.5)
plt.xlabel('Country',fontsize=25,color='yellow',fontstyle='italic',labelpad=15)
plt.ylabel('Gold Medals',fontsize=25,color='yellow',fontstyle='italic',labelpad=15)
plt.gca().set_facecolor('skyblue')
plt.gcf().set_facecolor('violet')
plt.xticks(color="#2F2FE4",fontsize=15)
plt.yticks(color="#2F2FE4",fontsize=15)
plt.tick_params(axis='x',pad=15)
plt.tick_params(axis='y',pad=15)
plt.grid(axis='y',color='black',linewidth=2)
plt.title('Bar chart',fontsize=25,color='yellow')
plt.show()
```

```python
#line plot
#since we already imported the libraries no need to import them again
data2=pd.DataFrame({
    'Day':['Mon','Tue','Wed','Thu','Fri'],
    'Sales':[20,35,30,45,50]
})
x=data2['Day']
y=data2['Sales']
plt.figure(figsize=(8,5))
plt.plot(x,y,marker='.',color='#471396',linestyle='dotted',linewidth=2,markersize=15)
plt.xlabel('DAYS',color='#2F578A',fontsize=25,fontstyle='oblique',labelpad=15)
plt.ylabel('SALES',color='#2F578A',fontsize=25,fontstyle='oblique',labelpad=15)
plt.xticks(color='#C3110C',fontsize=20,fontstyle='oblique')
plt.yticks(color='#C3110C',fontsize=10,fontstyle='oblique')
plt.gca().set_facecolor(color='#F68048')
plt.gcf().set_facecolor(color='#44444E')
plt.tick_params(axis='y',pad=8)
plt.tick_params(axis='x',pad=8)
plt.grid(axis='y')
plt.title('LinePlot',fontsize=25,color='#2F578A')
plt.show()
```

```python
#line chart 2 but same data
x=data2['Day']
y=data2['Sales']
plt.figure(figsize=(8,5))
plt.plot(x,y,marker='o',linestyle='dashed',color='green',markerfacecolor='yellow',markeredgecolor='black',markersize=10)
plt.title('Weekly Sales Report',color='cyan',fontsize=25)
plt.xlabel('Days',fontsize=25)
plt.ylabel('sales',fontsize=25)
plt.xticks(color='teal',fontsize=15)
plt.yticks(color='teal',fontsize=15)
plt.gca().set_facecolor(color='#2F2F2F')
plt.gcf().set_facecolor(color='#BFD7EA')
plt.grid(axis='y',color='black',alpha=0.6)
plt.show()
```
