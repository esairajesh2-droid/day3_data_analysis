```python
#importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\esair\OneDrive\Desktop\thelast\fifa_world_cup_2026_player_performance.csv")
```

```python
df_player=df[df['player_name']=='Lionel Martinez']
```

```python
player=df_player.sort_values(by='match_date',ascending=True).head(10)
print(player)
```

```python
#successfull passes against opponent teams
x=player.opponent_team
y=player.successful_passes
plt.figure(figsize=(8,5))
plt.title('Successful passes Against Opponent Teams',color='#FF00FF',fontsize=25,fontstyle='italic')
plt.bar(x,y,color='#00F5FF',width=0.8,edgecolor='#39FF14')
plt.xlabel('Opponent Team',fontsize=25,color='#FF00FF',fontstyle='italic',labelpad=15)
plt.ylabel('Successfull passes',color='#FF00FF',fontsize=25,fontstyle='italic',labelpad=15)
plt.xticks(rotation=90,color='#00FFFF',fontsize=10)
plt.yticks(color='#00FFFF',fontsize=10)
plt.tick_params(axis='x',pad=15)
plt.tick_params(axis='y',pad=15)
plt.gca().set_facecolor(color='#1A1A2E')
plt.gcf().set_facecolor(color='#0B0F1A')
plt.grid(axis='y',color='#5a1846',linewidth=2)
plt.tight_layout()
plt.show()

```

```python
#total minutes palyed against each opponent
x=player.opponent_team
y=player.minutes_played
plt.figure(figsize=(8,5))
plt.title("total minutes played by Lionel Martinez",color='#ffd54f',fontsize=25)
plt.plot(x,y,marker='o',markerfacecolor='#FFD54F',markeredgecolor='#000000',linestyle='solid',color='orange')
plt.xlabel('Opponent Team',color='#FFFFFF',fontsize=25,fontstyle='italic')
plt.ylabel('Minutes played',color='#FFFFFF',fontsize=25,fontstyle='italic')
plt.xticks(color='#ffd54f',fontsize=10,rotation=90)
plt.yticks(color='#ffd54f',fontsize=10,rotation=90)
plt.tick_params(axis='x',pad=8)
plt.tick_params(axis='y',pad=8)
plt.gca().set_facecolor(color='#0b0b0b')
plt.gcf().set_facecolor(color='#1c1c1c')
plt.grid(color='#444444',axis='y')
plt.tight_layout()
plt.show()
```

```python
#defensive contribution vs opponent
player=player.sort_values(by='defensive_contribution',ascending=False)
x=player.defensive_contribution
y=player.opponent_team
plt.figure(figsize=(8,5))
plt.barh(y,x,color='#F28C28')
plt.xlabel('Defensive Contribution',color='orange',fontsize=25,labelpad=8)
plt.ylabel('Opponent',color="orange",fontsize=25,labelpad=8)
plt.title('Defensive Contribution Vs Opponent',color='orange',fontsize=25)
plt.xticks(color='black',fontsize=15)
plt.yticks(color='black',fontsize=15)
plt.tick_params(axis='x',pad=8)
plt.tick_params(axis='y',pad=8)
plt.gca().set_facecolor(color='#FFF3CD')
plt.gcf().set_facecolor(color='#FFF8E1')
plt.grid(axis='x',linestyle="--",alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#Offensive contrinution vs opponent
x=player.offensive_contribution
y=player.opponent_team
plt.figure(figsize=(8,5))
plt.barh(y,x,color='#F28C28')
plt.xlabel('Offensive Contribution',color='orange',fontsize=25,labelpad=8)
plt.ylabel('Opponent',color="orange",fontsize=25,labelpad=8)
plt.title('Offensive Contribution Vs Opponent',color='orange',fontsize=25)
plt.xticks(color='black',fontsize=15)
plt.yticks(color='black',fontsize=15)
plt.tick_params(axis='x',pad=8)
plt.tick_params(axis='y',pad=8)
plt.gca().set_facecolor(color='#FFF3CD')
plt.gcf().set_facecolor(color='#FFF8E1')
plt.grid(axis='x',linestyle="--",alpha=0.3)
plt.tight_layout()
plt.show()
```
